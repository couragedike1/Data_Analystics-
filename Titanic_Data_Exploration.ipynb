ðŸ”¹ 1. Bar Chart: Passengers by Survival Status

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count passengers by survival status
survival_counts = df["Survived"].value_counts()

# Plot
plt.figure(figsize=(6, 4))
sns.barplot(x=survival_counts.index, y=survival_counts.values, palette=["red", "green"])

plt.xticks(ticks=[0, 1], labels=["Did Not Survive", "Survived"])
plt.ylabel("Number of Passengers")
plt.xlabel("Survival Status")
plt.title("Survival Count on Titanic")

plt.show()
```

ðŸ”¹ 2. Pie Chart: Passenger Class Distribution

```python
# Count passengers by class
class_counts = df["Pclass"].value_counts()

# Plot
plt.figure(figsize=(6, 6))
plt.pie(class_counts, labels=["3rd Class", "1st Class", "2nd Class"], autopct="%1.1f%%",
        colors=["blue", "gold", "green"], startangle=140)

plt.title("Titanic Passenger Class Distribution")
plt.show()
```

ðŸ”¹ 3. Histogram: Age Distribution of Passengers

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"].dropna(), bins=30, kde=True, color="purple")

plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.title("Age Distribution of Titanic Passengers")

plt.show()
```

ðŸ”¹ Find Passengers with "Jack" or "Rose" in Their Names


```python
# Filter passengers with names containing "Jack" or "Rose" (case-insensitive)
filtered_passengers = df[df["Name"].str.contains("Jack|Rose", case=False, na=False)]

# Display relevant details
print(filtered_passengers[["PassengerId", "Name", "Pclass", "Sex", "Age", "Fare", "Survived"]])
```

ðŸ”¹ Find the Youngest and Oldest Passenger
python


```python
# Find the youngest passenger
youngest_passenger = df[df["Age"] == df["Age"].min()][["PassengerId", "Name", "Pclass", "Sex", "Age", "Fare", "Survived"]]

# Find the oldest passenger
oldest_passenger = df[df["Age"] == df["Age"].max()][["PassengerId", "Name", "Pclass", "Sex", "Age", "Fare", "Survived"]]

# Print results
print("Youngest Passenger:\n", youngest_passenger)
print("\nOldest Passenger:\n", oldest_passenger)
```

:

ðŸ”¹ Find the Youngest and Oldest Passenger in pclass 1

```python
# Filter only first-class passengers
first_class_passengers = df[df["Pclass"] == 1]

# Find the youngest first-class passenger
youngest_first_class = first_class_passengers[first_class_passengers["Age"] == first_class_passengers["Age"].min()]

# Find the oldest first-class passenger
oldest_first_class = first_class_passengers[first_class_passengers["Age"] == first_class_passengers["Age"].max()]

# Display results
print("Youngest Passenger in First Class:\n", youngest_first_class[["PassengerId", "Name", "Pclass", "Sex", "Age", "Fare", "Survived"]])
print("\nOldest Passenger in First Class:\n", oldest_first_class[["PassengerId", "Name", "Pclass", "Sex", "Age", "Fare", "Survived"]])
```

```python
# Find Spedden, Master. Robert Douglas
child = df[df["Name"].str.contains("Spedden, Master. Robert Douglas", case=False, na=False)]

# Extract the last name "Spedden" (assuming family members share the last name)
last_name = "Spedden"

# Find other passengers with the same last name (potential parents)
family_members = df[df["Name"].str.contains(last_name, case=False, na=False)]

# Display relevant details
print("Spedden, Master. Robert Douglas:\n", child[["PassengerId", "Name", "Pclass", "Sex", "Age", "Survived"]])
print("\nSpedden Family Members:\n", family_members[["PassengerId", "Name", "Pclass", "Sex", "Age", "Survived"]])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Age"], y=df["Fare"], alpha=0.6, color="blue")

plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatter Plot of Age vs Fare")
plt.show()
```

 Correlation Between Age and Fare

```python
plt.figure(figsize=(8,5))
sns.regplot(x=df["Age"], y=df["Fare"], scatter_kws={"alpha":0.5}, line_kws={"color":"red"})

plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Regression Plot of Age vs Fare")
plt.show()
```

Compute Correlation by Passenger Class, Age and Fare

```python
# Compute correlation between Age and Fare for each Passenger Class
correlation_by_class = df.groupby("Pclass")[["Age", "Fare"]].corr().unstack().iloc[:, 1]

# Display correlation for each class
print("Correlation between Age and Fare by Passenger Class:\n", correlation_by_class)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(12, 5))

# Loop through each Pclass and plot regression line
for i, pclass in enumerate([1, 2, 3], 1):
    plt.subplot(1, 3, i)
    sns.regplot(x=df[df["Pclass"] == pclass]["Age"],
                y=df[df["Pclass"] == pclass]["Fare"],
                scatter_kws={"alpha": 0.5},
                line_kws={"color": "red"})
    plt.xlabel("Age")
    plt.ylabel("Fare")
    plt.title(f"Age vs Fare (Class {pclass})")

plt.tight_layout()
plt.show()
```
